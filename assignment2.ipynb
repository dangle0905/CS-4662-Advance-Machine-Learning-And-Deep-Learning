```python
#1- Download the Kears+Tensorflow tutorial from Canvas. Import all required modules including the following:
!pip install tensorflow
!pip install keras

from  keras.wrappers.scikit_learn  import  KerasClassifier
from  sklearn.model_selection  import  GridSearchCV
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
import numpy as np
```

```python
# Keras will download MNIST digit dataset for us:
from keras.datasets import mnist

#split into testing and training
(X_train, y_train), (X_test, y_test) = mnist.load_data()

#before reshaping
print(X_train.shape)
print(X_test.shape)


#2- Import the minst dataset, and split it into testing and training as we saw in the tutorial. Then, reshape each sample into a row vector, and scale it by dividing by 255.
# Reshape each image pixels into a row of feature table with 28*28=784 features (each pixel is a feature):
X_train = X_train.reshape(X_train.shape[0], 784)
X_test = X_test.reshape(X_test.shape[0], 784)


# simply scale the features to the range of [0,1]: by converting first
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255

# output label: 
print (y_train.shape)
#printing out the first 10 samples
print (y_train[:10])

print("this is X_train: ", X_train)
print("this is y_train: ", y_train)

```

```python
#3- Perform OneHotEncoding for the label y. So, your label will be a vector of 10 elements for each data sample.

# import some utilities to transform/preprocess our data: 
#our neural network will have 10 output
from keras.utils import np_utils
y_train = np_utils.to_categorical(y_train, 10)
y_test = np_utils.to_categorical(y_test, 10)

print (y_train.shape)
print (y_train[:10,:])
```

```python
#4- Now, define a function called model_creator. 
#This function will define, create, and compile your neural network model according to your structure, 
#and then return the built model as the output. For the ANN neurons/layers, 
#use the same structure as we had in the tutorial:

#the input are pixels
input_size = 784
#one hidden layer with 100 neurons
hidden_neurons = 100 
output_size = 10



def model_creator():
  # define:

    model = Sequential()

    # design the structure:
    
    #we are going to design the ANN Structure with 784 inputs, 10 outputs and 100 neuron in the hidden layer
    #---------------------------------------------------------------
    #first layer is the input layer
    #the input layers are simply just inputs and does not do any processing so we do not need to define the input layer

    #second layer is the hidden layer:

    model.add(Dense(hidden_neurons, input_dim = input_size))    
    model.add(Activation('sigmoid'))
    
    #third layer is the output layer:
    model.add(Dense(output_size, input_dim = hidden_neurons))
    model.add(Activation('softmax'))
    
   # compile:

    model.compile(loss='categorical_crossentropy', optimizer='adam',  metrics=['accuracy'])

   # return: 

    

    return model 
```

```python
#5- Fix the random state for reproducibility:     
seed = 2
np.random.seed(seed)
```

```python
#6- Use KerasClassifier class to wrap your model as an object:

model = KerasClassifier(build_fn = model_creator, verbose=2)
```

```python
#7- Now, run sklearn GridSearch to find the best batch_size and epochs. 
#Search in this range: batch_size = [30 , 50 , 100 ] , epochs = [10 , 15 , 20].  
#In your GridSearch, the estimator is the above model,  
#the scoring should be 'neg_log_loss', and you have to use 10-fold CV.
epochs = [10,15,20]
batch_size = [30,50,100]

from sklearn.model_selection import GridSearchCV

param_grid = dict(batch_size=batch_size, epochs=epochs)
gridTrain = GridSearchCV(model, param_grid, cv = 10, scoring = 'neg_log_loss')
gridTrain.fit(X_train, y_train)



```

```python
#8- Based on your results, what is the best batch_size and epochs?   
#Now, test your model with the best batch_size and epochs on the testing set.   grid.best_estimator_.model  gives you the best model found and trained in the grid-search. 
#What is the prediction accuracy on the testing set?

gridTest = GridSearchCV(model, param_grid, cv = 10, scoring = 'neg_log_loss')
gridTest.fit(X_test, y_test)

```

```python
print("The best batch size and epoch for training set is: ", gridTrain.best_params_)
print("The best batch size and epoch for testing set is: ", gridTest.best_params_)

print("The best model from training set: ", gridTrain.best_estimator_.model)
print("the best score from training set: ", gridTrain.best_score_)

print("The best model from testing set: ", gridTest.best_estimator_.model)
print("the best score from testing set: ", gridTest.best_score_)

# Prediction:
y_predict = gridTest.predict(X_test)


#our neural network will have 10 output
y_predict = np_utils.to_categorical(y_predict, 10)

#printing out some of the prediction results and test results
#print('y_predict', y_predict[:10])
#print('y_test', y_test[:10])

#importing to get accuracy score
from sklearn.metrics import accuracy_score

# Evaluation:
score = accuracy_score(y_test, y_predict)

print('The prediction accuracy on the testing set is: ', score)

```
