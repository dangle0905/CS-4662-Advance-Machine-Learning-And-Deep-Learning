```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from keras.layers import Dense, Dropout, Activation, Flatten
# import some utilities to transform our data
from keras.utils import np_utils
from tensorflow.keras import layers

#Write and submit your python codes in Jupyter Notebook to perform the following tasks.    

#Make sure to provide proper descriptions as MarkDown for each section of your code.  
 
#1- Deep Convolutional Autoencoder For Image Denoising: 
 
#In this problem, we use a very interesting and effective application of deep learning, specifically 
#Deep  Convolutional  Autoencoder.  We  implement  an  unsupervised  deep  convolutional 
#autoencoder for image de-noising, converting noisy digits images from the MNIST dataset to clear 
#digits images. (Make sure to review the Lab5 tutorial on CNN (in Canvas) before starting this HW). 
 
#a- Import all of the libraries and tools that you need including the following layers for your 
#network:  

from keras.layers import Convolution2D, MaxPooling2D, Conv2DTranspose 

#b Import the mnist dataset and split it into testing and training: 
from keras.datasets import mnist 
(X_train, y_train), (X_test, y_test) = mnist.load_data() 

#reshape the X_train and X_test to correct dimensions
X_train = X_train.reshape(X_train.shape[0], 28, 28, 1)
X_test = X_test.reshape(X_test.shape[0], 28, 28, 1)

# simply normalize the features to the range of [0,1]:
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
X_train /= 255
X_test /= 255

# c- Use the following function to add Random noise to all your training and testing samples, and 
#save them as new datasets noisy_X_train and noisy_X_test. 

def noise(array): 
    noise_factor = 0.4 
    noisy_array = array + noise_factor * np.random.normal( 
            loc=0.0, scale=1.0, size=array.shape) 
    return np.clip(noisy_array, 0.0, 1.0) 

noisy_X_train = noise(X_train);
noisy_X_test = noise(X_test);


#d- Plot some of your samples with noise and without noise and compare them. You should see 
#something like this: 
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

print("no noise from X_train")
imgplot = plt.imshow(X_train[9,:,:], cmap='viridis')
plt.show()

print("noise from noisy_X_train")
imgplot = plt.imshow(noisy_X_train[9,:,:], cmap='viridis')
plt.show()

print("no noise from X_test")
imgplot = plt.imshow(X_test[9,:,:],cmap='viridis')
plt.show()

print("noise from noisy_X_test")
imgplot = plt.imshow(noisy_X_test[9,:,:], cmap='viridis')
plt.show()


#e- Now, we want to create our deep convolutional autoencoder for de-noising the images. As 
#you saw in class lectures, the autoencoder includes an Encoder part that compresses the 
#input, and a Decoder part that reconstruct the data.  
#Add the following layers in order: 

# "Sequential" model lets us to define a stack of neural network layers
from keras.models import Sequential
# Declare Sequential model fro our network:
model = Sequential()

# import the "core" layers:
from keras.layers import Dense, Dropout, Activation, Flatten

# Encoder:
model.add(Convolution2D(32, (3, 3), activation="relu", padding="same")) 
model.add(MaxPooling2D((2, 2), padding="same")) 
model.add(Convolution2D(32, (3, 3), activation="relu", padding="same") )
model.add(MaxPooling2D((2, 2), padding="same"))

# Decoder: 
model.add(Conv2DTranspose(32, (3, 3), strides=2, activation="relu", padding="same") )
model.add(Conv2DTranspose(32, (3, 3), strides=2, activation="relu", padding="same") )
model.add(Convolution2D(1, (3, 3), activation="sigmoid", padding="same") )

#Note: Notice the Conv2DTranspose layer, which is called Transposed convolution also known 
#as  Deconvolution.  Dont  forget  the  strides=2  (the  step  size  of  2).  Also,  notice  the  last 
#Convolution2D layer that is used to convert the tensor to the same size as we had in the input. 

#f- Compile the model with  loss='binary_crossentropy', optimizer='adam' 
model.compile(loss='binary_crossentropy', optimizer='adam')
model.summary

#g- Train the model ONLY using the training data. Remember, you dont use y (label) here. Your 
#input  will  be  noisy_X_train  and  your  output  will  be  X_train  (reconstruction).  Use  the 
#batch_size=128 and epochs=5:
model.fit(noisy_X_train, X_train, batch_size=128, epochs=5) 

#h- Now, Make prediction on your noisy testing set to de-noise it. The output of your prediction 
#will be the de-noised version of your noisy testing set: 
#predict(noisy_X_test) 
prediction = model.predict(noisy_X_test)
print("The denoised results")
imgplot = plt.imshow(prediction[3,:,:])
plt.show()

#i- It is time to see the Magic! Plot 10 random images of your testing set before and after de-
#noising to see how amazing your model is in cleaning noisy images. You should see something 
#like this: 

for n in range(9):
    print("noise")
    imgplot = plt.imshow(noisy_X_test[n,:,:])
    plt.show()
    print("de-noise")
    imgplot = plt.imshow(prediction[n,:,:])
    plt.show()
```
